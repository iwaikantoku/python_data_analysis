```python
def create_dataset(num):
    data_x = np.linspace(0,1,num)
    data_z = np.sin(2*np.pi*data_x)
    data_z1 = normal(0,0.3,num)
    data_y = data_z + data_z1
    return data_y
```

```python
import pandas as pd
def create_dataset(num):
    data_x = np.linspace(0,1,num)
    data_z = np.sin(2*np.pi*data_x)
    data_z1 = normal(0,0.3,num)
    data_y = data_z + data_z1
    df = pd.DataFrame({'X':data_x, 'Y':data_y})
    return df
```

```python
df = create_dataset(1000)
```

```python
df.describe()
```

```python
data = create_dataset(10)    
```

```python
def coin_game(money, bet):
    coin = randint(2)
    if coin == 0:
        money += bet
    else:
        money -= bet
    return money
```

```python
money = 1000
trial1 = list()
for i in range(100):
   
    money = coin_game(money, money/2)
    trial1.append(money)
   
```

```python
money = 1000
trial2 = list()
for i in range(100):
   
    money = coin_game(money, money/2)
    trial2.append(money)
```

```python
money = 1000
trial3 = list()
for i in range(100):
   
    money = coin_game(money, money/2)
    trial3.append(money)
```

```python
df = pd.DataFrame({'trial0':trial1,'trial1':trial2,'trial2':trial3})
plt.plot(df)
plt.show()
```

```python
def stock_game(money, bet):
    coin = randint(2)
    if coin == 0:
        money += bet
    else:
        money -= bet/2
    return money
```

```python
money = 1000
trial1 = list()
for i in range(100):
   
    money = stock_game(money, money/2)
    trial1.append(money)
   
```

```python
money = 1000
trial2 = list()
for i in range(100):
   
    money = stock_game(money, money/2)
    trial2.append(money)
   
```

```python
money = 1000
trial3 = list()
for i in range(100):
   
    money = stock_game(money, money/2)
    trial3.append(money)
   
```

```python
df = pd.DataFrame({'trial0':trial1,'trial1':trial2,'trial2':trial3})
plt.plot(df)
plt.show()
```
